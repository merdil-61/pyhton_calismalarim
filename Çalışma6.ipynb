```python
for sayi in range(1,11):
  if sayi==3:
    continue #3'ü atla
  if sayi==7:
    break #7'de döngüyü durdur
  print(sayi)
```

0'dan 99'a kadar olan tüm sayıları ekrana yazdıran programı yazınız.

```python
sayi=0
toplam=0
while sayi<99:
  print(sayi)
  toplam+=sayi
  sayi+=1
print("Sayıların toplamaı:", toplam)
```

Büyük harfler işle alfabedeki harfleri ekrana yazdıran programı yazınız.

```python
for ch in range(ord('A'),ord('Z')+1):
  print("Büyük Harfler:", chr(ch))
```

Girilen bir sayının pozitif mi negatif mi olduğunu bulan programın kodunu yazınız

```python
while True:
  sayi=float(input("Bir sayı girin:"))

  if sayi==0:
    print("Sayı sıfırdır.Lütgfen 0 dışında bir sayı girin")
    continue
  elif sayi>0:
    print("Sayı pozitiftir.")
  else:
    print("Sayı negatiftir.")
  break
```

1 ile 999 arasında yer alan tamsayıların toplamını yazdıran programı yazınız.


```python
toplam =0
sayi=1
while sayi<=999:
  toplam+=sayi
  sayi+=1
  print(toplam)
```

1'den N'e kadar tamsayıların toplamını yazdıran programı yazınız.



```python
n=int(input("bir sayı girin:"))
toplam=0
for sayac in range(1,n+1):
  toplam+=sayac
print("Sonuc:" , toplam) #Matematiksel işlemde tümevarım metodu--Toplam=n*(n+1)/2
```
